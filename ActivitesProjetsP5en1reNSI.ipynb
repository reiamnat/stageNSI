````python
from random import randint

bouge = False
coordonnees = [300, 150]

def setup():    
    createCanvas(600, 200)
    noStroke()

def draw():    
    etape1(mouseX, mouseY)   
    etape2(mouseX, mouseY)  
    etape3(mouseX, mouseY)
    etape4()
        

#---------------------------------------------------------------------
def etape1(x, y):
    background(220)
    if survole_zone(x, y):
        dessine_boutons_vides()
        dessine_motifs_boutons()
    else :
        dessine_boutons_vides()
        

def etape2(x, y):
    if survole_play(x, y):
        dessine_play_couleur()
    elif survole_pause(x, y):
        dessine_pause_couleur()
    elif survole_stop(x, y):
        dessine_stop_couleur()  

        
def etape3(x, y):
    global bouge
    if (mouseIsPressed and survole_play(x, y)):
        dessine_play_couleur_click()
        bouge = True
    elif (mouseIsPressed and survole_pause(x, y)):
        dessine_pause_couleur_click()
        bouge = False
    elif (mouseIsPressed and survole_stop(x, y)):
        dessine_stop_couleur_click()
        bouge = False
        noLoop()
        
def etape4():
    if bouge:
        coordonnees[0] = coordonnees[0] + randint(-3, 3)
    fill('red')
    circle(coordonnees[0], coordonnees[1], 15)
               
#-----------------------------------------------------------------------
def survole_zone(x,y):
    if 0 <= x <= 300 and 0 <= y <= 100:
        fill('lightgray')
        rect(0,0, 300, 100)
        return True

def dessine_boutons_vides():
    fill(150)
    circle(75, 50, 50)
    circle(150, 50, 50)
    circle(225, 50, 50)
    
def dessine_motifs_boutons():
    fill(50)
    triangle (70, 40, 70, 60, 85, 50)
    rect(143, 40, 5, 20)
    rect(152, 40, 5, 20)
    square(215, 40, 20)

def distance(x1, y1, x2, y2):
    return sqrt((x2 - x1)**2 + (y2 - y1)**2)
    
def survole_play(x, y):
    return distance(x, y, 75, 50) < 25

def survole_pause(x, y):
    return distance(x, y, 150, 50) < 25

def survole_stop(x, y):
    return distance(x, y, 225, 50) < 25

def dessine_play_couleur():
    fill('green')
    circle(75, 50, 50)
    fill('white')
    triangle (70, 40, 70, 60, 85, 50)

def dessine_play_couleur_click():
    fill('green')
    circle(75, 50, 55)
    fill('lightgreen')
    triangle (70, 40, 70, 60, 85, 50)

def dessine_pause_couleur():
    fill('orange')
    circle(150, 50, 50)
    fill('White')
    rect(143, 40, 5, 20)
    rect(152, 40, 5, 20)

def dessine_pause_couleur_click():
    fill('orange')
    circle(150, 50, 55)
    fill('orangeRed')
    rect(143, 40, 5, 20)
    rect(152, 40, 5, 20)
    
def dessine_stop_couleur():
    fill('red')
    circle(225, 50, 50)
    fill('White')
    square(215, 40, 20)

def dessine_stop_couleur_click():
    fill('red')
    circle(225, 50, 55)
    fill('FireBrick')
    square(215, 40, 20)
#---------------------------------------------------------------------------

run()


````

# 4. Un automate cellulaire à une dimension.
    
Le parcours suivant permet de mettre en œuvre les connaissances acquises sur les tableaux, et de construire un tableau de tableaux.
    
</div>

<div class = "alert alert-block alert-success">
    
### Exemple : la règle 126.

On envisage les huit cas possibles pour une cellule et ses deux voisines, et on détermine l'état suivant de la cellule centrale.
    
<img src = https://raw.githubusercontent.com/reiamnat/stageNSI/main/capture-decran-2021-11-01-a-18.59.13.png width=500>
    
Comment lire ce tableau ?   
- Dans la première colonne : une cellule vivante, entourée de deux cellules vivantes, meurt à l'étape suivante.   
- Colonnes 2, 5 et 6 : une cellule vivante qui possède une seule ou aucune voisine vivante, reste vivante à l'étape suivante.   
- Colonnes 3, 4 et 7 : une cellule morte qui possède une ou deux voisine(s) vivante(s), devient vivante à l'étape suivante.   
- Dernière colonne :  une cellule morte qui possède deux voisines mortes, reste morte à l'étape suivante.  
    
Cette règle est appelée **règle 126** car si l'on associe les cellules vivantes à un 1 et les cellules mortes à un 0, la dernière ligne du tableau représente l'entier 0111 1110, soit 126 en binaire. 
Cette règle pourrait être défénie à l'aide du tableau suivant : 
    
|111|110|101|100|011|010|001|000|
|:-:|:-:|:-:|:-:|:-:|:-:|:-:|:-:|
|0|1|1|1|1|1|1|0|

En supposant que l'on a comme population initiale une unique cellule vivante située au centre, comment évolue la population ?
<img src=https://raw.githubusercontent.com/reiamnat/stageNSI/main/gen_126_0.png width=300>

Voici les deux "étapes" suivantes : 
    <img src=https://raw.githubusercontent.com/reiamnat/stageNSI/main/gen_126_1.png width=300>

<img src=https://raw.githubusercontent.com/reiamnat/stageNSI/main/gen_126_2.png width=300>

On représente les différentes générations les unes sous les autres : 
<img src=https://raw.githubusercontent.com/reiamnat/stageNSI/main/gen_126_0_8.png width = 300>

</div>


````python
# correction
def population_initiale(n):
    '''
    n est un entier impair
    population_initiale(n) est un tableau de n valeurs égales à 0, 
    sauf la valeur centrale qui vaut 1. 
    '''
    ligne = [0] * n
    ligne[(n - 1) // 2] = 1    #l'indice n // 2 convient aussi puisque n est impair
    return ligne


def suivante_126(ligne):
    '''
    ligne est un tableau de 0 et 1, représentant une population de cellules
    suivante_126(ligne) est le tableau représentant la population ligne à l'étape suivante,
    en respectant la règle 126
    '''
    suivante = [0] * len(ligne)
    suivante[0] = ligne[0]
    for i in range(1, len(ligne) - 1):
        if ligne[i - 1] == ligne[i] and  ligne[i] == ligne[i + 1]:
            suivante[i] = 0
        else:
            suivante[i] = 1
    suivante[len(ligne) - 1] = ligne[len(ligne) - 1]
    return suivante


def grille_evolution_126(n, nb_generations):
    '''
    n est un entier représentant la taille de la population de chaque génération
    nb_generations est un entier représentant le nombre de générations à engendrer
    grille_evolution_126(n, nb_generations) est un tableau de tableaux : la population initiale 
    comporte n cellules dont seule la cellule centrale est vivante, et chacune des nb_generations populations 
    suivantes est obtenue en appliquant la règle 126 à la population de la génération précédente.
    '''
    populations =  [population_initiale(n)]
    for i in range(1, nb_generations):
        populations.append(suivante_126(populations[i - 1]))
    return populations


def dessine_grille(grille, taille_cellule):
    for i in range(len(grille)):         # i est l'indice de ligne
        for j in range(len(grille[0])):  # j est l'indice de colonne
            if grille[i][j] == 0:
                fill(255)
            else:
                fill(0)
            square(j * taille_cellule , i * taille_cellule, taille_cellule)
````

````python
N = 195
TAILLE_CELLULE = 5

def setup():
    createCanvas(TAILLE_CELLULE * N, TAILLE_CELLULE * N // 2)
    stroke(220)
    noLoop()
    
def draw():
    populations = grille_evolution_126(N, N // 2)
    dessine_grille(populations, TAILLE_CELLULE)
        
run()
````

# 5. Une deuxième animation
    
Le parcours suivant permet de construire une liste de points, chacun étant représenté par un dictionnaire.
    
</div>

````python
# correction complète
from random import randint, choice

points = [] 

def points_sur_cercle(x, y, r):
    for i in range(0, 360, 10):
        x_sur_cercle = int(x + r * cos(i))    # la fonction int permet d'obtenir un arrondi entier
        y_sur_cercle = int(x + r * sin(i))
        
        # Ajout d'un nouveau point-cible à la liste points
        points.append({'x_cible':x_sur_cercle, 'y_cible':y_sur_cercle})
        
def distribuer_position_aleatoire():
    for point in points:
        point['x'] = randint(0, 400)
        point['y'] = randint(0, 400)

def distribuer_couleur():
    couleurs = ["DarkSalmon", "IndianRed", "FireBrick", "Crimson", "Orange",
            'LightSalmon', 'Tomato', 'OrangeRed', 'DarkOrange']
    for point in points:
        point['couleur'] = choice(couleurs)
        
def bouger():
    for point in points:
        if point['x'] < point['x_cible']:
            point['x'] = point['x'] + 1
        elif point['x'] > point['x_cible']:
            point['x'] = point['x'] - 1

        if point['y'] < point['y_cible']:
            point['y'] = point['y'] + 1
        elif point['y'] > point['y_cible']:
            point['y'] = point['y'] - 1

def afficher():
    for point in points:
        fill(point['couleur'])
        circle(point['x'], point['y'], 10)

def setup():
    createCanvas(400, 400)
    noStroke()
    angleMode(DEGREES)  
    points_sur_cercle(200, 200, 150)
    distribuer_position_aleatoire()
    distribuer_couleur()

def draw():
    background(220, 100)  
    bouger()
    afficher()
        
run()
````

# 6. Un automate cellulaire à deux dimensions : la fourmi de Langton
    
    
</div>

````python
def grille_initiale(n):
    '''
    renvoie une grille carrée de taille n * n, ne contenant que des zéros
    '''
    grille = [[0] * n for i in range(n)]
    return grille

def dessiner_grille_vide(nb_cases, dim_case):
    '''
    dessine la grille initiale : grille de nb_cases lignes et nb_cases colonnes
    Chaque case est un carré de dimension dim_case.
    '''
    strokeWeight(1)  # épaisseur du tracé
    stroke(180)      # couleur de tracé (gris)
    fill(255)        # couleur de remplissage (blanc)
    for i in range(nb_cases):
        for j in range(nb_cases):
            square(i * dim_case, j * dim_case, dim_case)

def fourmi_initiale(n):
    '''
    renvoie le dictionnaire dont les clés sont 'ligne', 'colonne' et 'direction' pour une fourmi située
    au centre d'une grille de taille n * n, orientée vers le nord
    '''
    return {'ligne':n // 2, 'colonne':n // 2, 'direction':3}
   
def diriger_fourmi(fourmi):
    '''
    fourmi est un tableau de 3 valeurs représentant la position (ligne, colonne) et la direction de la fourmi.
    diriger_fourmi modifie la direction de la fourmi en fonction de la couleur de la case sur laquelle elle se trouve.
    '''
    if  grille[fourmi['ligne']][fourmi['colonne']] == 0: # test sur la couleur de la case courante
        fourmi['direction'] = (fourmi['direction'] + 1) % 4    # changement de direction : à droite   
    else:
        fourmi['direction'] = (fourmi['direction'] - 1) % 4    # changement de direction : à gauche
    
def bouger_fourmi(fourmi):
    '''
    fourmi est un tableau de 3 valeurs représentant la position (ligne, colonne) et la direction de la fourmi.
    bouger_fourmi modifie la position de la fourmi en fonction de sa direction.
    '''
    if fourmi['direction'] == 0:            # direction vers l'est
        fourmi['colonne'] = fourmi['colonne'] + 1 # déplacement d'une colonne vers l'est
    elif fourmi['direction'] == 1:          # direction vers le sud
        fourmi['ligne'] = fourmi['ligne'] + 1 # déplacement d'une ligne vers le sud
    elif fourmi['direction'] == 2:          # direction vers l'ouest
        fourmi['colonne'] = fourmi['colonne'] - 1 # déplacement d'une colonne vers l'ouest
    else:                         # direction vers le nord
        fourmi['ligne'] = fourmi['ligne'] - 1 # déplacement d'une ligne vers le nord   
    
def tracer_case(ligne, colonne, couleur, dim_case):
    '''
    tracer_case(ligne, colonne, couleur, dim_case) trace à la position (ligne, colonne) 
    une case de la couleur spécifiée, dans une grille dont les cases ont pour dimension dim_case.
    '''
    stroke(180)
    fill(couleur)
    square(colonne * dim_case, ligne * dim_case, dim_case)

def inverser_couleur(ligne, colonne, dim_case):
    '''
    inverser_couleur(ligne, colonne, dim_case) modifie la couleur associée à la case repérée par 
    ses indices de ligne et colonne dans la variable grille 
    et appelle la fonction tracer_case pour afficher cette modification à l'écran
    '''
    if  grille[ligne][colonne] == 0: # test sur couleur de la case 
        grille[ligne][colonne] = 1   # inversion de couleur   
        couleur = 0                  # niveau de gris de la couleur ou nom de la couleur
    else:
        grille[ligne][colonne] = 0   # inversion de couleur 
        couleur = 255                # niveau de gris de la couleur ou nom de la couleur
    tracer_case(ligne, colonne, couleur, dim_case)   
    
def est_dans_grille(ligne, colonne, grille):
    '''
    est_dans_grille(ligne, colonne, grille) renvoie True si les indices de ligne et colonne 
    correspondent à une case de grille et False sinon.
    '''
    if ligne >= 0 and colonne >= 0 and ligne < len(grille) and colonne < len(grille):
        return True
    else:
        noLoop()
        return False

def tester_nb_etapes(nb_etapes_max):
    '''
    tester_nb_etapes(nb_etapes_max) interrompt le programme si le nombre d'étapes réalisées 
    est supérieur à nb_etapes_max
    '''
    if frameCount >= nb_etapes_max:
        noLoop()
        
def dessiner_fourmi(dim_case):
    '''
    dessine une "fourmi" dans une case de dimension dim_case
    l'origine du repère est le centre de la case
    '''
    pas = dim_case // 8
    # les pattes et antennes
    line(0, 2*pas, 0, -2*pas)
    line(-2*pas, -2*pas, -pas // 3,  0)
    line(-2*pas,  2*pas, -pas // 3,  0)
    line( 2*pas, -2*pas,  pas // 3,  0)
    line( 2*pas,  2*pas,  pas // 3,  0)
    line(2 * pas, 0, 3 * pas, -pas)
    line(2 * pas, 0, 3 * pas,  pas)
    # le corps
    noStroke()
    ellipse(0,  0, dim_case //3, pas)
    ellipse(-2 * pas,  0, 2 * pas, dim_case // 6)
    ellipse(2 * pas, 0, dim_case // 6, dim_case // 6 )

def tracer_fourmi(fourmi, dim_case):
    '''
    fourmi est un dictionnaire dont les clés sont `ligne`, `colonne`et `direction`. 
    tracer_fourmi(fourmi, dim_case) trace une "fourmi" dans une case de dimension dim_case,
    à la position et dans la direction contenues dans le dictionnaire fourmi
    '''
    fill(220, 30, 30)
    stroke('red')
    push()
    translate(fourmi['colonne'] * dim_case + dim_case // 2, fourmi['ligne'] * dim_case + dim_case // 2)
    rotate(fourmi['direction'] * PI/2)
    dessiner_fourmi(dim_case)
    pop()

def afficher_nb_etapes(x, y):
    '''
    affiche le nombre frameCount à la position (x, y)
    '''
    textSize(36)
    fill(255)
    noStroke()
    rect(x, y - textAscent(frameCount), textWidth(frameCount), 2 * textAscent(frameCount))
    fill(0)
    text(frameCount, x, y)

#--------------------------------------------------------------------------------------

NB_CASES = 10                     # nombre de cases du quadrillage par ligne ou colonne
FRAMERATE = 1                   # nombre d'images par seconde
NB_ETAPES_MAX = 12000

#--------------------------------------------------------------------------------------


DIM_CASE = 800 // NB_CASES        # dimension d'une case
grille = grille_initiale(NB_CASES)
fourmi = fourmi_initiale(NB_CASES)

def setup():
    createCanvas(800, 800)
    dessiner_grille_vide(NB_CASES, DIM_CASE)
    tracer_fourmi(fourmi, DIM_CASE)
    frameRate(FRAMERATE)
    
def draw():
    diriger_fourmi(fourmi)
    inverser_couleur(fourmi['ligne'], fourmi['colonne'], DIM_CASE)
    bouger_fourmi(fourmi)
    if est_dans_grille(fourmi['ligne'], fourmi['colonne'], grille):
        tracer_fourmi(fourmi, DIM_CASE)
    tester_nb_etapes(NB_ETAPES_MAX)
    afficher_nb_etapes(DIM_CASE, NB_CASES * DIM_CASE)
    
#--------------------------------------------------------------------------------------
run()

````

````python
def affiche_grille(tab):
    """
    tab est une grille contenant des 0 et des 1
    affiche_grille(tab) affiche sous forme de grille les données de tab : 
    les 0 sont représentés par des cases vertes et les 1 par des cases bleues
    """
    lignes = len(tab)                    # nombre de lignes
    colonnes = len(tab[0])               # nombre de colonnes
    taille = 400 / max(lignes, colonnes) # dimension d'une case
  
    for ligne in range(lignes):
        for colonne  in range(colonnes):
            if tab[ligne][colonne] == 0:
                couleur = color(0, 148, 133) # vert
            elif tab[ligne][colonne] == 1:
                couleur = color(0, 104, 218) # bleu
            fill(couleur)
            square(colonne * taille, ligne * taille, taille) # colonne renseigne sur l'abscisse, et ligne sur l'ordonnée
            
def partie_gagnee(grille):
    for ligne in grille :
        for element in ligne :
            if element == 0 :
                return False
    return True

def inverse_valeur(grille, ligne, colonne):
    grille[ligne][colonne] = 1 - grille[ligne][colonne]
    # permet d'échanger les 0 en 1 et les 1 en 0


def quelle_case_cliquee(x, y, longueur_grille):
    """
    renvoie les indices de ligne et de colonne dans une grille carrée de la case qui contient 
    le pixel de coordonnées (x,y) 
    """
    # taille d'une case, en pixels
    taille_case = 400 // len(grille) 
    if 0 <= x <= width and 0 <= y <= height:
        return (y // taille_case, x // taille_case)

def voisins(ligne, colonne, dimension_grille):
    liste_voisins = []
    
    #gauche 
    if colonne > 0 :
        liste_voisins.append((ligne, colonne - 1))
        
    #droite 
    if colonne < dimension_grille - 1 :
        liste_voisins.append((ligne, colonne + 1))
    
    #haut 
    if ligne > 0 :
        liste_voisins.append((ligne - 1, colonne))
        
    #bas 
    if ligne < dimension_grille - 1 :
        liste_voisins.append((ligne + 1, colonne))
    
    return liste_voisins

              
def souris_cliquee(grille):
    global clic_en_cours_de_traitement, nb_coups
    if mouseIsPressed:
        if clic_en_cours_de_traitement == False:
            clic_en_cours_de_traitement = True
            
            ligne, colonne = quelle_case_cliquee(int(mouseX), int(mouseY), len(grille))
            inverse_valeur(grille, ligne, colonne)
        
            for voisine in voisins(ligne, colonne, len(grille)):
                ligne_case, colonne_case = voisine
                inverse_valeur(grille, ligne_case, colonne_case)
            
            nb_coups = nb_coups + 1
    
    else: # si la souris n'est plus pressée
        clic_en_cours_de_traitement = False                            

def finale(nb_coups):
    affiche_grille(grille)
    fill(0, 148, 133)
    rect(40, 150, 320, 50)
    textSize(24)
    text("BRAVO : victoire en "+str(nb_coups) +" coups", 50, 180)
````
