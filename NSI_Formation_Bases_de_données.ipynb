## Base "films-cnam.sql" : 

Cinémathèque proposée par le CNAM (http://deptfod.cnam.fr/bd/tp) regroupant 6 tables : 
    
**Film** (<u>idFilm</u>, titre, année, #idRéalisateur, genre, résumé, #codePays)

**Pays** (<u>code</u>, nom, langue)

**Artiste** (<u>idArtiste</u>, nom, prénom, annéeNaiss)

**Rôle** (<u>idFilm, idActeur</u>, nomRôle)

**Internaute** (<u>email</u>, nom, prénom, #codePays)

**Notation** (<u>idFilm, email</u>, note)
