## Exemple : 
</div>

```python
# réglages initiaux
bgcolor("LightBlue")   # couleur du fond
pencolor("navy")       # couleur des tracés
fillcolor("SteelBlue") # couleur de remplissage
penup()
goto(-280, -30)
pendown()

# polygones réguliers
for nb_cotes in range(3, 7):
    begin_fill()
    for cote in range(nb_cotes):
        fd(80)
        left(360 // nb_cotes)
    end_fill()
    if nb_cotes < 6:
        penup()
        fd(50 + nb_cotes * 25)
        pendown()
    
hideturtle()
done()
```
